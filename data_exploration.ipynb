A partir da matriz de correlaÃ§Ã£o, podemos identificar as seguintes observaÃ§Ãµes e insights:

A variÃ¡vel cutoff_factor apresenta correlaÃ§Ã£o alta com actual_distance_to_destination e actual_time. Isso sugere que o fator de corte pode estar influenciando diretamente o tempo e a distÃ¢ncia das viagens, talvez indicando que viagens mais longas e com maior distÃ¢ncia possam ter maior fator de corte.
3. CorrelaÃ§Ã£o do trip_duration com Outros Tempos:
trip_duration (duraÃ§Ã£o da viagem) tem correlaÃ§Ã£o alta com variÃ¡veis como actual_time, osrm_time, e actual_distance_to_destination. Isso indica que a duraÃ§Ã£o total da viagem estÃ¡, como esperado, fortemente ligada aos tempos e distÃ¢ncias estimados.
4. VariÃ¡veis com CorrelaÃ§Ã£o Baixa (segment_factor e factor):
segment_factor e factor tÃªm correlaÃ§Ãµes baixas ou mesmo negativas com vÃ¡rias outras variÃ¡veis, o que pode indicar que eles tÃªm uma influÃªncia menos direta nos tempos e nas distÃ¢ncias da viagem, ou que representam fatores mais independentes, possivelmente relacionados a condiÃ§Ãµes especÃ­ficas de segmentos da rota ou caracterÃ­sticas externas.
5. RelaÃ§Ã£o entre Segmentos e Totais de Tempo e DistÃ¢ncia:
As variÃ¡veis relacionadas aos segmentos de rota, como segment_actual_time, segment_osrm_time, e segment_osrm_distance, tÃªm correlaÃ§Ã£o positiva com actual_time e osrm_time, mas de forma moderada. Isso sugere que os tempos e distÃ¢ncias dos segmentos contribuem para o total, mas podem variar em relaÃ§Ã£o ao tempo e Ã  distÃ¢ncia totais.
6. CorrelaÃ§Ã£o entre trip_duration e time_to_cutoff:
Existe uma correlaÃ§Ã£o positiva considerÃ¡vel entre trip_duration e time_to_cutoff (0.80). Isso sugere que, quanto maior a duraÃ§Ã£o da viagem, maior o tempo atÃ© o corte, indicando que viagens mais longas podem estar mais alinhadas com o tempo de corte.
ConclusÃµes Potenciais:
Esses insights sugerem que algumas variÃ¡veis-chave, como distÃ¢ncia e tempo, estÃ£o fortemente interligadas, e variÃ¡veis como cutoff_factor e segment_factor podem representar caracterÃ­sticas adicionais de controle. As relaÃ§Ãµes fortes podem indicar redundÃ¢ncia entre variÃ¡veis, o que pode ser considerado ao selecionar features para um modelo de machine learning.

```python
#H1
# Plotar o boxplot do tempo de viagem por tipo de rota
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='route_type', y='actual_time')
plt.title('DistribuiÃ§Ã£o do Tempo de Viagem por Tipo de Rota')
plt.xlabel('Tipo de Rota')
plt.ylabel('Tempo de Viagem (actual_time)')
plt.show()
```

```python
#H1
# EstatÃ­sticas descritivas do tempo de viagem por tipo de rota
route_stats = df.groupby('route_type')['actual_time'].describe()
print(route_stats)
```

```python
#H1
from scipy.stats import f_oneway, kruskal

# Separar os tempos de viagem por tipo de rota
groups = [group["actual_time"].values for name, group in df.groupby("route_type")]

# ANOVA para dados normalmente distribuÃ­dos
f_stat, p_value_anova = f_oneway(*groups)
print("ANOVA - F-statistic:", f_stat, "P-value:", p_value_anova)

# Kruskal-Wallis para dados nÃ£o normais
h_stat, p_value_kruskal = kruskal(*groups)
print("Kruskal-Wallis - H-statistic:", h_stat, "P-value:", p_value_kruskal)
```

```python
#H2
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='route_type', y='actual_time')
plt.title('DistribuiÃ§Ã£o do Tempo de Entrega por Tipo de Rota')
plt.xlabel('Tipo de Rota')
plt.ylabel('Tempo Real de Entrega')
plt.show()


# Separar os tempos de entrega por tipo de rota
groups = [group["actual_time"].values for name, group in df.groupby("route_type")]

# ANOVA
f_stat, p_value_anova = f_oneway(*groups)
print("ANOVA - F-statistic:", f_stat, "P-value:", p_value_anova)

# Kruskal-Wallis
h_stat, p_value_kruskal = kruskal(*groups)
print("Kruskal-Wallis - H-statistic:", h_stat, "P-value:", p_value_kruskal)
```

```python
#H3
# Converter para datetime se necessÃ¡rio
df['trip_creation_time'] = pd.to_datetime(df['trip_creation_time'])

# Criar coluna de horÃ¡rio de pico
df['peak_hour'] = df['trip_creation_time'].dt.hour.isin([7, 8, 9, 17, 18, 19])

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='peak_hour', y='actual_time')
plt.title('DistribuiÃ§Ã£o do Tempo de Entrega em HorÃ¡rios de Pico e Fora de Pico')
plt.xlabel('HorÃ¡rio de Pico')
plt.ylabel('Tempo Real de Entrega')
plt.xticks([0, 1], ['Fora de Pico', 'Pico'])
plt.show()

from scipy.stats import ttest_ind, mannwhitneyu

# Separar os tempos de entrega em horÃ¡rio de pico e fora de pico
peak_times = df[df['peak_hour']]['actual_time']
off_peak_times = df[~df['peak_hour']]['actual_time']

# Teste T
t_stat, p_value_ttest = ttest_ind(peak_times, off_peak_times)
print("Teste T - T-statistic:", t_stat, "P-value:", p_value_ttest)

# Teste Mann-Whitney
u_stat, p_value_mannwhitney = mannwhitneyu(peak_times, off_peak_times)
print("Mann-Whitney - U-statistic:", u_stat, "P-value:", p_value_mannwhitney)
```
