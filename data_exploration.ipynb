A partir da matriz de correlação, podemos identificar as seguintes observações e insights:

1. Correlação Forte entre Distância e Tempo:
As variáveis actual_distance_to_destination, osrm_distance, actual_time, e osrm_time têm uma correlação muito alta entre si (próxima de 1), o que indica que o tempo e a distância de viagem estão fortemente relacionados. Isso faz sentido, pois viagens mais longas tendem a ter tempos maiores.
2. Alta Correlação com cutoff_factor:
A variável cutoff_factor apresenta correlação alta com actual_distance_to_destination e actual_time. Isso sugere que o fator de corte pode estar influenciando diretamente o tempo e a distância das viagens, talvez indicando que viagens mais longas e com maior distância possam ter maior fator de corte.
3. Correlação do trip_duration com Outros Tempos:
trip_duration (duração da viagem) tem correlação alta com variáveis como actual_time, osrm_time, e actual_distance_to_destination. Isso indica que a duração total da viagem está, como esperado, fortemente ligada aos tempos e distâncias estimados.
4. Variáveis com Correlação Baixa (segment_factor e factor):
segment_factor e factor têm correlações baixas ou mesmo negativas com várias outras variáveis, o que pode indicar que eles têm uma influência menos direta nos tempos e nas distâncias da viagem, ou que representam fatores mais independentes, possivelmente relacionados a condições específicas de segmentos da rota ou características externas.
5. Relação entre Segmentos e Totais de Tempo e Distância:
As variáveis relacionadas aos segmentos de rota, como segment_actual_time, segment_osrm_time, e segment_osrm_distance, têm correlação positiva com actual_time e osrm_time, mas de forma moderada. Isso sugere que os tempos e distâncias dos segmentos contribuem para o total, mas podem variar em relação ao tempo e à distância totais.
6. Correlação entre trip_duration e time_to_cutoff:
Existe uma correlação positiva considerável entre trip_duration e time_to_cutoff (0.80). Isso sugere que, quanto maior a duração da viagem, maior o tempo até o corte, indicando que viagens mais longas podem estar mais alinhadas com o tempo de corte.
Conclusões Potenciais:
Esses insights sugerem que algumas variáveis-chave, como distância e tempo, estão fortemente interligadas, e variáveis como cutoff_factor e segment_factor podem representar características adicionais de controle. As relações fortes podem indicar redundância entre variáveis, o que pode ser considerado ao selecionar features para um modelo de machine learning.

```python
#H1
# Plotar o boxplot do tempo de viagem por tipo de rota
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='route_type', y='actual_time')
plt.title('Distribuição do Tempo de Viagem por Tipo de Rota')
plt.xlabel('Tipo de Rota')
plt.ylabel('Tempo de Viagem (actual_time)')
plt.show()
```

```python
#H1
# Estatísticas descritivas do tempo de viagem por tipo de rota
route_stats = df.groupby('route_type')['actual_time'].describe()
print(route_stats)
```

```python
#H1
from scipy.stats import f_oneway, kruskal

# Separar os tempos de viagem por tipo de rota
groups = [group["actual_time"].values for name, group in df.groupby("route_type")]

# ANOVA para dados normalmente distribuídos
f_stat, p_value_anova = f_oneway(*groups)
print("ANOVA - F-statistic:", f_stat, "P-value:", p_value_anova)

# Kruskal-Wallis para dados não normais
h_stat, p_value_kruskal = kruskal(*groups)
print("Kruskal-Wallis - H-statistic:", h_stat, "P-value:", p_value_kruskal)
```

```python
#H2
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='route_type', y='actual_time')
plt.title('Distribuição do Tempo de Entrega por Tipo de Rota')
plt.xlabel('Tipo de Rota')
plt.ylabel('Tempo Real de Entrega')
plt.show()


# Separar os tempos de entrega por tipo de rota
groups = [group["actual_time"].values for name, group in df.groupby("route_type")]

# ANOVA
f_stat, p_value_anova = f_oneway(*groups)
print("ANOVA - F-statistic:", f_stat, "P-value:", p_value_anova)

# Kruskal-Wallis
h_stat, p_value_kruskal = kruskal(*groups)
print("Kruskal-Wallis - H-statistic:", h_stat, "P-value:", p_value_kruskal)
```

```python
#H3
# Converter para datetime se necessário
df['trip_creation_time'] = pd.to_datetime(df['trip_creation_time'])

# Criar coluna de horário de pico
df['peak_hour'] = df['trip_creation_time'].dt.hour.isin([7, 8, 9, 17, 18, 19])

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='peak_hour', y='actual_time')
plt.title('Distribuição do Tempo de Entrega em Horários de Pico e Fora de Pico')
plt.xlabel('Horário de Pico')
plt.ylabel('Tempo Real de Entrega')
plt.xticks([0, 1], ['Fora de Pico', 'Pico'])
plt.show()

from scipy.stats import ttest_ind, mannwhitneyu

# Separar os tempos de entrega em horário de pico e fora de pico
peak_times = df[df['peak_hour']]['actual_time']
off_peak_times = df[~df['peak_hour']]['actual_time']

# Teste T
t_stat, p_value_ttest = ttest_ind(peak_times, off_peak_times)
print("Teste T - T-statistic:", t_stat, "P-value:", p_value_ttest)

# Teste Mann-Whitney
u_stat, p_value_mannwhitney = mannwhitneyu(peak_times, off_peak_times)
print("Mann-Whitney - U-statistic:", u_stat, "P-value:", p_value_mannwhitney)
```
